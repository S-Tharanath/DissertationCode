```python
import requests
def fetch_and_clean(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            clean_content = remove_tags(response.content)
            return clean_content
        else:
            print(f"Failed to fetch content from {url}. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"An error occurred while fetching content from {url}: {str(e)}")
        return None
def remove_tags(html):
    soup = BeautifulSoup(html, "html.parser")
    for data in soup(['style', 'script']):
        data.decompose()
    return ' '.join(soup.stripped_strings)
urls = ['https://dopefashionsense.com/2023/09/coachs-spring-2024-ready-to-wear-collection-review/','https://hypebae.com/2023/9/spring-summer-2024-new-york-fashion-week-collina-strada','https://www.harpersbazaar.com/fashion/designers/a45050784/proenza-schouler-spring-2024-collection-skirt/','https://hypebeast.com/2023/9/palomo-spain-ss24-spring-summer-2024-collection-new-york-fashion-week-runway','https://hypebae.com/2023/9/kim-shui-spring-summer-2024-new-york-fashion-week','https://hypebeast.com/2023/9/dion-lee-ss24-spring-summer-2024-collection-new-york-fashion-week-runway','https://theimpression.com/michael-kors-spring-2024-fashion-show-review/','https://www.elle.com/runway/a45105686/brandon-maxwell-spring-2024-review/','https://www.fashionotography.com/tory-burch-spring-summer-2024-nyfw/','https://kendam.com/news/fashion-shows/laquan-smith-spring-summer-2024-runway','https://www.fashionotography.com/jason-wu-collection-spring-summer-2024/','https://www.vogue.com/fashion-shows/spring-2024-ready-to-wear/prabal-gurung','https://www.vogue.com/fashion-shows/spring-2024-ready-to-wear/altuzarra','https://www.vogue.com/fashion-shows/spring-2024-ready-to-wear/helmut-lang','https://hypebeast.com/2023/9/helmut-lang-ss24-spring-summer-2024-collection-new-york-fashion-week-runway','https://hypebeast.com/2023/9/ralph-lauren-ss24-spring-summer-2024-collection-new-york-fashion-week-runway','https://hypebeast.com/2023/9/peter-do-spring-summer-2024-paris-fashion-week-runways','https://hypebeast.com/2023/9/saint-laurent-spring-summer-2024-paris-fashion-week-womenswear-runway','https://fuckingyoung.es/loewe-spring-summer-2024/','https://hypebeast.com/2023/9/marni-spring-summer-2024-paris-fashion-week-runway','https://umsquaremagazine.com/fashion/e1wqifd50fqz0wdc4l0oa9a0mvfxpl','https://wwd.com/runway/spring-2024/paris/dries-van-noten/review/','https://hypebeast.com/2023/10/valentino-spring-summer-2024-paris-fashion-week-menswear-womenswear-runway','https://www.fashionotography.com/rabanne-spring-summer-2024-pfw/','https://www.fashionotography.com/chloe-spring-2024-paris-fashion-week/','https://theimpression.com/givenchy-spring-2024-fashion-show-review/','https://wwd.com/runway/spring-2024/paris/isabel-marant/review/','https://hypebeast.com/2023/9/alexander-mcqueen-spring-summer-2024-paris-fashion-week-menswear-womenswear-runway','https://wwd.com/runway/spring-2024/paris/miu-miu/review/#!','https://theimpression.com/miu-miu-spring-2024-fashion-show-review/','https://www.thezoereport.com/fashion/chanel-spring-summer-2024-review','https://theimpression.com/louis-vuitton-spring-2024-fashion-show-review/','https://kendam.com/news/fashion-shows/alberta-ferretti-spring-summer-2024-runway','https://www.vogue.com/fashion-shows/spring-2024-ready-to-wear/del-core','https://www.vogue.com/fashion-shows/spring-2024-ready-to-wear/fendi','https://theimpression.com/gucci-spring-2024-fashion-show-review/','https://theimpression.com/prada-spring-2024-fashion-show-review/','https://www.thezoereport.com/fashion/prada-spring-summer-2024-review','https://theimpression.com/versace-spring-2024-fashion-show-review/','https://hypebeast.com/2023/9/tom-ford-spring-summer-2024-milan-fashion-week-runway','https://hypebeast.com/2023/9/ferragamo-spring-summer-2024-menswear-womenswear-milan-fashion-week-runway','https://www.vogue.com/fashion-shows/spring-2024-ready-to-wear/david-koma','https://www.fashionotography.com/16arlington-spring-summer-2024-lfw/','https://hypebeast.com/2023/9/burberry-spring-summer-2024-menswear-womenswear-london-fashion-week-runway-daniel-lee','https://www.harpersbazaar.com/fashion/fashion-week/a45224620/supriya-lele-london-fashion-week-spring-2024/','https://www.thezoereport.com/fashion/simone-rocha-spring-summer-2024-review','https://www.fashionotography.com/molly-goddard-spring-summer-2024-lfw/','https://vprcommag.com/2023/09/london-fashion-week-ss24-di-petsa/','https://www.vogue.com/fashion-shows/spring-2024-ready-to-wear/charlotte-knowles#review']
aggregated_data = ""
for url in urls:
    cleaned_content = fetch_and_clean(url)
    if cleaned_content:
        aggregated_data += cleaned_content + " "
print(aggregated_data)
```

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
text = aggregated_data
doc = nlp(text)
```

```python
dependency_tree = [(token.text, token.dep_, token.head.text) for token in doc]
print("Dependency Parsing:", dependency_tree)
```

```python
trends_in_collection = []
target_dependencies = ["amod","compound","pobj"]
```

```python
for token in doc:
    if token.pos_ in ["ADJ","NOUN"] and token.dep_ in target_dependencies:
        phrase = token.text
        for child in token.subtree:
            if child.pos_ in ["ADJ","NOUN"] and child.dep_ in target_dependencies and child.text != phrase:
                phrase += " " + child.text
        trends_in_collection.append(phrase)
```

```python
colors = ["red", "blue", "green", "yellow", "orange", "purple", "pink", "brown", "black", "white", "gray", "beige","ecru","gold","silver","metallic","light blue","pastel","lilac","cherry red","maroon","off-white","navy","olive green","turquoise","tiffany blue","coral","orange","teal","lime green","iridescent","mauve","taupe","indigo","neutral","copper","lavender","khaki","neutrals","blush","ivory","scarlet","cerise","burgundy","cream"]
col_syn = {'blush':'neutrals','neutral':'neutrals','gold':'metallic','silver':'metallic','metallics':'metallic','ecru':'neutrals','beige':'neutrals','off-white':'white','ivory':'white','lavender':'lilac','scarlet':'red','cerise':'red'}
color_trends =[]
import re
for phrase in trends_in_collection:
    # Check if the phrase matches any color in the color list using regular expressions
    if any(re.search(r'\b{}\b'.format(color), phrase, re.IGNORECASE) for color in colors):
        color_trends.append(phrase)
std_color_list = [col_syn.get(phrase,phrase) for phrase in color_trends]
print("\nColor Trends:", std_color_list)
```

```python
sil_synonyms = {'loose':'relaxed','skinny':'fitted','voluminous':'oversized','bustier':'corsets','bodycon':'fitted','corset':'corsets'}
silhouettes = ["sleek", "voluminous", "a-line", "ballgown", "asymmetric", "hobble", "shift", "slip", "corset", "oversized", "relaxed","wide leg","skinny","bodycon","bootcut","flare","mini","midi","maxi","pencil","draped","loose","fitted","layered","circle","bias cut","tube","ruched","slim","full-length","knee-length","corsets","flared"]
shapes = []
for phrase in trends_in_collection:
    if any(re.search(r'\b{}\b'.format(silhouette), phrase, re.IGNORECASE) for silhouette in silhouettes):
        shapes.append(phrase)
std_sil_length_list = [sil_synonyms.get(phrase, phrase) for phrase in shapes]
print("\nSilhouette Trends:", std_sil_length_list)
```

```python
materials = ["silk", "satin", "organza", "tweed", "lightweight", "cotton", "wool", "tulle", "chiffon", "sheer","velvet", "poplin","lace","leather","fur","faux fur","faux leather","denim","recycled","chain","metal","feather","nylon","linen","jacquard","knitted","suede","mohair","transparent","see-through","mesh","crepe"]
mat_synonyms = {'see-through':'sheer','transparent':'sheer','see through':'sheer','knitted':'knit','knit':'knit'}
# Categorize fashion trends into various categories
fabrics = []
for phrase in trends_in_collection:
    if any(re.search(r'\b{}\b'.format(material), phrase, re.IGNORECASE) for material in materials):
        fabrics.append(phrase)
fabric_list = [mat_synonyms.get(phrase,phrase) for phrase in fabrics]
print("\nFabric Trends:", fabric_list)
```

```python
prints = ["striped", "checkered", "flannel", "polka-dots","dots","floral", "paisley", "abstract", "leopard", "cheetah", "animal print", "spots","ruffle" ,"plaid","glitter","shiny","textured","snake","crocodile","solid","monochrome","monochrome with accent","argyle","tribal","monogram","logo","camoflage","fringe","fringed","flowers","nature","pleats","pleated","ruffled","ruffles","sequins","sequinned","shimmer","frills","monochromatic","minimalistic","minimalism","cargo","slit","slits","cutout","cutouts"]
pat_syn = {"monochrome":"monochromatic","flannel":"plaid","dot":"polka-dots","cheetah":"animal print","leopard":"animal print","flowers":"floral","sequinned":"sequins","pleats":"pleated","fringed":"fringe"}
# Categorize fashion trends into various categories
patterns = []
for phrase in trends_in_collection:
    if any(re.search(r'\b{}\b'.format(print), phrase, re.IGNORECASE) for print in prints):
        patterns.append(phrase)
pattern_list = [pat_syn.get(phrase,phrase) for phrase in patterns]
print("\nFabric Trends:", pattern_list)
```

```python
import matplotlib.pyplot as plt
from collections import Counter
color_count = Counter()
for phrase in std_color_list:
    cols = phrase.split()  # Split into individual words
    for col in cols:
        if col in colors:
            color_count[col] += 1
colors1 = list(color_count.keys())
frequencies = list(color_count.values())
plt.figure(figsize=(25, 5))
plt.bar(colors1, frequencies, color='black')
plt.xlabel('Colors')
plt.ylabel('Frequency')
plt.title('Color trends in Spring/Summer 2024 Fashion Weeks')
plt.show()
```

```python
sil_length_count = Counter()
for phrase in std_sil_length_list:
    sils = phrase.split()  
    for sil in sils:
        if sil in silhouettes:
            sil_length_count[sil] += 1
silhouettes1 = list(sil_length_count.keys())
frequencies = list(sil_length_count.values())
plt.figure(figsize=(25, 5))
plt.bar(silhouettes1, frequencies, color='pink')
plt.xlabel('Silhouettes and lengths')
plt.ylabel('Frequency')
plt.title('Silhouette and length trends in Spring/Summer 2024 Fashion Weeks')
plt.show()
```

```python
fabric_count = Counter()
for phrase in fabric_list:
    fabs = phrase.split()  # Split into individual words
    for fab in fabs:
        if fab in materials:
            fabric_count[fab] += 1
fabrics1 = list(fabric_count.keys())
frequencies = list(fabric_count.values())
plt.figure(figsize=(25, 5))
plt.bar(fabrics1, frequencies, color='black')
plt.xlabel('Fabrics')
plt.ylabel('Frequency')
plt.title('Fabric trends in Spring/Summer 2024 Fashion Weeks')
plt.show()
```

```python
pat_count = Counter()
for phrase in pattern_list:
    pats = phrase.split()  # Split into individual words
    for pat in pats:
        if pat in prints:
            pat_count[pat] += 1
patterns1 = list(pat_count.keys())
frequencies = list(pat_count.values())
plt.figure(figsize=(25, 5))
plt.bar(patterns1, frequencies, color='pink')
plt.xlabel('Patterns')
plt.ylabel('Frequency')
plt.title('Embellishment trends in Spring/Summer 2024 Fashion Weeks')
plt.show()
```
